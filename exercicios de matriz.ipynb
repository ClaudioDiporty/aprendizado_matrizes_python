### Fun√ß√µes do Programa Principal

```python
def exibir_matriz(matriz): 
    for linha in matriz:
        print(linha) 
    return ''
def soma_matrix (A, B):
    C = []
    nLinhasA, nLinhasB = len(A), len(B)
    nColA, nColB = len(A[0]), len(B[0])
    
    if (nLinhasA == nLinhasB) and (nColA == nColB):
        for i in range(nLinhasA):
            linha = [0]*nColA
            C.append(linha)
            for j in range(nColA):
                C[i][j] = A[i][j] + B[i][j]       
    else:
        print('Matrizes n√£o tem a mesma ordem')    
    return C
def oposta(M):
    op = []
    for i in range(len(M)):
        linha =[0]*len(M[0])
        op.append(linha)
        for j in range(len(M[0])):
            op[i][j] = -M[i][j]
    return op
def subtrair(A, B):
    return soma_matrix(A, oposta(B))    
```
